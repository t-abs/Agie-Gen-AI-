```python
from crewai.tools import Tool, tool_decorator

@tool_decorator(name="WeatherTool", description="Fetches current weather for a given location")
def get_current_weather(location: str) -> str:
    import requests
    API_KEY = 'da96be69-37bc-4878-816c-fed2d429cf0a'
    url = f"http://api.weatherapi.com/v1/current.json?key={API_KEY}&q={location}"
    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()
        weather = data['current']['condition']['text']
        temperature = data['current']['temp_c']
        return f"The current weather in {location} is {weather} with a temperature of {temperature}°C."
    else:
        return f"Failed to fetch weather data for {location}."

# Register the tool in CrewAI
weather_tool = Tool(name="WeatherTool", func=get_current_weather)

# Example usage
location = "London"
print(weather_tool.execute(location))
```

```python
import crewai
from crewai.agents import Agent
from crewai.tasks import Task

# Define the Data Analyst agent
data_analyst_agent = Agent(
    role='Data Analyst',
    goal='analyzing sales data',
    backstory='an experienced data analyst with a focus on sales trends',
    tools=['data visualization']
)

# Define the 'collect sales data' task for the Data Analyst
collect_sales_data_task = Task(
    description="Collect sales data for analysis",
    agent=data_analyst_agent,
    expected_output="A detailed report containing the collected sales data",
    context="Sales data for the last quarter"
)

# Add the task to the Data Analyst agent
data_analyst_agent.add_task(collect_sales_data_task)

# Define the Manager agent
manager_agent = Agent(
    role='Manager',
    goal='overseeing data collection and analysis',
    backstory='an experienced manager with a focus on optimizing data processes',
    tools=[]
)

# Define a task for the Manager to delegate 'collect sales data' task to the Data Analyst
delegate_task = Task(
    description="Delegate the task of collecting sales data to the Data Analyst",
    agent=manager_agent,
    expected_output="The Data Analyst agent has been assigned the task of collecting sales data",
    context="Delegating task to Data Analyst"
)

# Add the delegation task to the Manager agent
manager_agent.add_task(delegate_task)

# Delegate the task
def delegate_collect_sales_data_task():
    # Manager assigns the 'collect sales data' task to the Data Analyst
    data_analyst_agent.start()

# Register the delegation function as a tool for the Manager
manager_agent.add_tool(
    name="DelegateCollectSalesDataTask",
    func=delegate_collect_sales_data_task,
    description="Delegates the task of collecting sales data to the Data Analyst"
)

# Run the Manager agent to perform the delegation
manager_agent.start()

print("Manager agent has delegated the task to Data Analyst agent.")
```

```python
import crewai
from crewai.agents import Agent
from crewai.tasks import Task

# Define the callback function
def log_output_callback(step, output):
    with open("process_log.txt", "a") as log_file:
        log_file.write(f"Step: {step}, Output: {output}\n")

# Define the Data Analyst agent
data_analyst_agent = Agent(
    role='Data Analyst',
    goal='analyzing sales data',
    backstory='an experienced data analyst with a focus on sales trends',
    tools=['data visualization'],
    callback=log_output_callback  # Register the callback function
)

# Define the 'collect sales data' task for the Data Analyst
collect_sales_data_task = Task(
    description="Collect sales data for analysis",
    agent=data_analyst_agent,
    expected_output="A detailed report containing the collected sales data",
    context="Sales data for the last quarter"
)

# Add the task to the Data Analyst agent
data_analyst_agent.add_task(collect_sales_data_task)

# Define the Manager agent
manager_agent = Agent(
    role='Manager',
    goal='overseeing data collection and analysis',
    backstory='an experienced manager with a focus on optimizing data processes',
    tools=[],
    callback=log_output_callback  # Register the callback function
)

# Define a task for the Manager to delegate 'collect sales data' task to the Data Analyst
delegate_task = Task(
    description="Delegate the task of collecting sales data to the Data Analyst",
    agent=manager_agent,
    expected_output="The Data Analyst agent has been assigned the task of collecting sales data",
    context="Delegating task to Data Analyst"
)

# Add the delegation task to the Manager agent
manager_agent.add_task(delegate_task)

# Delegate the task
def delegate_collect_sales_data_task():
    # Log the delegation step
    log_output_callback("Delegation Step", "Manager is delegating the task to Data Analyst")
    # Manager assigns the 'collect sales data' task to the Data Analyst
    data_analyst_agent.start()

# Register the delegation function as a tool for the Manager
manager_agent.add_tool(
    name="DelegateCollectSalesDataTask",
    func=delegate_collect_sales_data_task,
    description="Delegates the task of collecting sales data to the Data Analyst"
)

# Run the Manager agent to perform the delegation
manager_agent.start()

print("Manager agent has delegated the task to Data Analyst agent.")
```

```python
import crewai
from crewai.agents import Agent

def create_agents():
    agents = []

    # Define agent roles, goals, and backstories
    roles_goals_backstories = [
        {
            "role": "Researcher",
            "goal": "conducting thorough research on assigned topics",
            "backstory": "an experienced researcher with a background in data analysis and academic studies"
        },
        {
            "role": "Writer",
            "goal": "crafting engaging and informative content",
            "backstory": "a talented writer with a passion for storytelling and a knack for clear communication"
        },
        {
            "role": "Editor",
            "goal": "reviewing and refining written content to ensure quality and coherence",
            "backstory": "a meticulous editor with an eye for detail and a deep understanding of language and style"
        }
    ]

    # Create agents based on the defined roles, goals, and backstories
    for rgb in roles_goals_backstories:
        agent = Agent(
            role=rgb["role"],
            goal=rgb["goal"],
            backstory=rgb["backstory"],
            tools=[]  # Add tools as needed
        )
        agents.append(agent)

    return agents

# Example usage
agents = create_agents()

# Print the details of each agent to verify creation
for agent in agents:
    print(f"Role: {agent.role}, Goal: {agent.goal}, Backstory: {agent.backstory}")
```

```python
import crewai
from crewai.agents import Agent
from crewai.tasks import Task

# Define the callback function to log output to a file
def log_output_callback(step, output):
    with open("process_log.txt", "a") as log_file:
        log_file.write(f"Step: {step}, Output: {output}\n")

# Define the Content Creator agent
content_creator_agent = Agent(
    role='Content Creator',
    goal='creating engaging and informative articles',
    backstory='a creative writer with a knack for storytelling and content creation',
    tools=[],  # Add tools as needed
    callback=log_output_callback
)

# Define the Proofreader agent
proofreader_agent = Agent(
    role='Proofreader',
    goal='reviewing and refining written content for accuracy and clarity',
    backstory='a meticulous proofreader with an eye for detail and language precision',
    tools=[],  # Add tools as needed
    callback=log_output_callback
)

# Define the task for creating an article
create_article_task = Task(
    description="Create an engaging and informative article",
    agent=content_creator_agent,
    expected_output="A completed article draft",
    context="Writing a new article"
)

# Define the task for proofreading an article
proofread_article_task = Task(
    description="Proofread the article for accuracy and clarity",
    agent=proofreader_agent,
    expected_output="A proofread article with corrections",
    context="Reviewing the article draft"
)

# Add the tasks to the respective agents
content_creator_agent.add_task(create_article_task)
proofreader_agent.add_task(proofread_article_task)

# Function to alternate between Content Creator and Proofreader agents
def alternating_process():
    # Content Creator creates an article
    log_output_callback("Creation Step", "Content Creator is creating an article")
    content_creator_agent.start()

    # Proofreader reviews the article
    log_output_callback("Proofreading Step", "Proofreader is reviewing the article")
    proofreader_agent.start()

# Register the alternating process as a tool for each agent (if needed)
content_creator_agent.add_tool(
    name="AlternateProcess",
    func=alternating_process,
    description="Alternates between Content Creator and Proofreader to complete articles"
)

proofreader_agent.add_tool(
    name="AlternateProcess",
    func=alternating_process,
    description="Alternates between Content Creator and Proofreader to complete articles"
)

# Run the alternating process
alternating_process()

print("Alternating process between Content Creator and Proofreader has started.")
```

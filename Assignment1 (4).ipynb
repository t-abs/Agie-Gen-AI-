```python
import fitz  # PyMuPDF

# Function to extract text from PDF
def extract_text_from_pdf(pdf_path):
    document = fitz.open(pdf_path)
    text_content = ""

    for page_num in range(len(document)):
        page = document.load_page(page_num)
        text_content += page.get_text()

    return text_content

# Path to the PDF file
pdf_path = '/content/English-Bhagavad-gita-His-Divine-Grace-AC-Bhaktivedanta-Swami-Prabhupada.pdf'

# Extract text from the PDF
text_content = extract_text_from_pdf(pdf_path)

# Write the extracted text to a file
with open('extracted_text.txt', 'w', encoding='utf-8') as text_file:
    text_file.write(text_content)

print("Text content extracted and saved to 'extracted_text.txt'")
```

```python

from sentence_transformers import SentenceTransformer
import faiss
import numpy as np

# Function to split text into chunks
def split_text(text, chunk_size=512):
    words = text.split()
    return [' '.join(words[i:i + chunk_size]) for i in range(0, len(words), chunk_size)]

# Load the extracted text
with open('extracted_text.txt', 'r', encoding='utf-8') as file:
    text_content = file.read()

# Split the text into chunks
chunks = split_text(text_content)

# Initialize the Sentence Transformer model
model = SentenceTransformer('paraphrase-MiniLM-L6-v2')

# Encode the chunks
embeddings = model.encode(chunks)

# Create a FAISS index
index = faiss.IndexFlatL2(embeddings.shape[1])
index.add(np.array(embeddings))

# Save the index and the chunks
faiss.write_index(index, 'faiss_index.bin')
np.save('text_chunks.npy', chunks)

print("Text indexed and saved.")
```

```python
# Load the index and chunks
index = faiss.read_index('faiss_index.bin')
chunks = np.load('text_chunks.npy', allow_pickle=True)

# Function to search the index
def search_index(query, model, index, chunks, k=5):
    query_embedding = model.encode([query])
    distances, indices = index.search(np.array(query_embedding), k)
    return [chunks[i] for i in indices[0]]

# Example query
query = "Sri Shankaracharya"
results = search_index(query, model, index, chunks)

print("Search results for query:", query)
for result in results:
    print(result)
    print("------")
```

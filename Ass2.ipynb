```python
!pip install llama-index scikit-learn matplotlib

from llama_index.llms.mistralai import MistralAI
from llama_index.embeddings.mistralai import MistralAIEmbedding
import numpy as np
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
from scipy.spatial.distance import cdist

api_key = "j5O2Zp5maOchwRRN6VCwtNvVyQSJbNzf"

# Initialize MistralAI LLM and embedding model
llm = MistralAI(api_key=api_key)
resp = llm.complete("Hello Tanisha, nice to meet you!")
print(resp)

embedding_model_name = "mistral-embed"
embed_model = MistralAIEmbedding(model_name=embedding_model_name, api_key=api_key)
embedding = embed_model.get_text_embedding("Hello Krishna, nice to meet you!")
print(f"Dimension of embeddings: {len(embedding)}")

# Function to get embeddings for a list of texts
def get_embeddings(texts):
    return [embed_model.get_text_embedding(text) for text in texts]

# Define text inputs
texts = [
    "The quick brown fox jumps over the lazy dog.",
    "A fast auburn fox leaps over a dormant canine.",
    "The rain in Spain stays mainly in the plain.",
    "Computers and smartphones are ubiquitous in modern life."
]

# Generate embeddings for text inputs
embeddings = get_embeddings(texts)
embeddings = np.array(embeddings)

# Calculate cosine distances
distances = cdist(embeddings, embeddings, 'cosine')
print("Cosine distances between embeddings:\n", distances)

# Define batch text inputs
batch_texts = [
    "Artificial intelligence is transforming the world.",
    "Machine learning algorithms can find patterns in data.",
    "Natural language processing allows computers to understand human language."
]

# Generate embeddings for batch text inputs
batch_embeddings = get_embeddings(batch_texts)
batch_embeddings = np.array(batch_embeddings)

# Combine all embeddings
all_embeddings = np.vstack([embeddings, batch_embeddings])

# Use t-SNE for dimensionality reduction
tsne = TSNE(n_components=2, perplexity=5, random_state=42)  # Adjust perplexity here
reduced_embeddings = tsne.fit_transform(all_embeddings)

# Plot the embeddings
plt.figure(figsize=(10, 7))
plt.scatter(reduced_embeddings[:len(texts), 0], reduced_embeddings[:len(texts), 1], color='blue', label='Original Texts')
plt.scatter(reduced_embeddings[len(texts):, 0], reduced_embeddings[len(texts):, 1], color='green', label='Batch Texts')
for i, text in enumerate(texts + batch_texts):
    plt.annotate(text, (reduced_embeddings[i, 0], reduced_embeddings[i, 1]))

plt.legend()
plt.title('t-SNE Visualization of Text Embeddings')
plt.xlabel('Dimension 1')
plt.ylabel('Dimension 2')
plt.show()
```
